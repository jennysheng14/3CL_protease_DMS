```python
sets_res = sf.sets_and_residues(spreadsheet)
sets_res = [(x[1], x[0]) for x in sets_res]
test_dict = dict(sets_res)
```

### Activity/Resistance/Resistance for GRL data

```python
# activity

activity = pd.read_csv('/home/jennysheng/Yeast/raw_value_matrices_nosyn/Glu_Gal.csv', index_col = 0)
activity_std = pd.read_csv('/home/jennysheng/Yeast/raw_value_matrices_std_nosyn/Glu_Gal.csv', index_col = 0)
act_melt = pd.melt(activity.reset_index(), id_vars='index')
std_melt = pd.melt(activity_std.reset_index(), id_vars='index')
merge_melt = act_melt.merge(std_melt, on = ['index', 'variable'])
merge_melt['set'] = merge_melt['variable'].apply(lambda x: test_dict[x])
merge_melt.columns = ['amino_acid', 'residue', 'activity',
                     'std', 'set']
merge_melt.to_csv('activity_score_by_set.csv')
```

```python
fig = px.scatter(x = merge_melt['activity'], y = merge_melt['std'],
          color = merge_melt['set'])
fig.update_layout(
        xaxis_title="Score",
        yaxis_title="Standard deviation",
        autosize=False,
        width=900,
        height=500,
        font=dict(family="Arial",
                        size=16,
                        color="#020202"
                        ))
fig.show()
fig.write_html(fig_folder+"activity_score_by_set.html")
```
